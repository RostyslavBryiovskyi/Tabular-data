```python
# Supress Warnings

import warnings
warnings.filterwarnings('ignore')

# Import the numpy and pandas package

import numpy as np
import pandas as pd

# Data Visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

**Read the data**

```python
dataset = pd.read_csv("train.csv")
dataset.head()
```

**Data cleaning**

```python
# Checking Null values
dataset.isnull().sum()*100/dataset.shape[0]
```

**Exploratory Data Analysis**

```python
# Explore the range of target, its outliers and quantiles
plt.figure(figsize=(9, 8))
sns.boxplot(dataset['target'])
plt.show()
```

```python
# Explore the correlation between all variables.
plt.figure(figsize=(25, 25))
corr = round(dataset.corr(),2)
sns.heatmap(corr, cmap="YlGnBu", annot = True)
plt.show()
```

```python
# Now, let`s see the pairwise plots between each input feature and target
for i in range(0, len(dataset.columns), 5):
    sns.pairplot(data=dataset,
                x_vars=dataset.columns[i:i+5],
                y_vars=['8'])
```

**After these steps, we can conclude that feature number 6 is the only significant one, most correlated with the target (positive correlation = 0.9), in an inverse normal distribution."**
